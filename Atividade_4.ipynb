```python
import random
import pandas as pd

frutas = ["maçã", "banana", "laranja", "uva", "maçã", "melão", "mamão", "banana"]

with open('Minhas_frutas.txt', 'w') as file:
    for fruta in frutas:
        quantidade = random.randint(0, 100)
        file.write(f"{fruta},{quantidade}\n")

frutas_quantidades = {}

with open('Minhas_frutas.txt', 'r') as file:
    for line in file:
        nome, quantidade = line.strip().split(',')
        quantidade = int(quantidade)

        if nome in frutas_quantidades:
            frutas_quantidades[nome] += quantidade
        else:
            frutas_quantidades[nome] = quantidade

df = pd.DataFrame(list(frutas_quantidades.items()), columns=["Fruta", "Quantidade"])

print(df)
```
