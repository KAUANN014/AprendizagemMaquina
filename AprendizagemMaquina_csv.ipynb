```python
import csv
from datetime import datetime

# Dados para o CSV
data = data = data = [
    ("João Silva", "05/12/1990", "2025-02-24", "14:30:00"),
    ("Maria Oliveira", "15/02/1985", "2025-02-24", "15:00:00"),
    ("Carlos Souza", "20/08/1992", "2025-02-24", "09:45:00"),
    ("Ana Costa", "11/10/1995", "2025-02-24", "11:15:00"),
    ("Pedro Almeida", "30/03/1988", "2025-02-24", "17:00:00"),
]

file_name = "/content/cadastro.csv"

with open(file_name, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Escrevendo o cabeçalho
    writer.writerow(["Nome", "Data de Nascimento", "Dia de Cadastro", "Hora de Cadastro"])
    # Escrevendo os dados
    writer.writerows(data)

print(f"Arquivo CSV {file_name} criado com sucesso!")
```

```python
import csv

# Lendo o arquivo CSV
with open("/content/cadastro.csv", mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # Ignorar o cabeçalho
    registros = [row for row in reader]

print("Arquivo CSV lido com sucesso!")
```

```python
# Solicita o número do registro
registro_n = int(input("Digite o número do registro (1 a 5): "))

# Verifica se o número do registro é válido
if 1 <= registro_n <= len(registros):
    nome, data_nascimento, dia_cadastro, hora_cadastro = registros[registro_n - 1]

    dia_cadastro_obj = datetime.strptime(dia_cadastro, "%Y-%m-%d")
    dia_cadastro_formatado = dia_cadastro_obj.strftime("%d/%m/%Y")

    # Imprime o registro solicitado
    print(f"{nome} {data_nascimento} {dia_cadastro_formatado} {hora_cadastro}")
else:
    print("Número do registro inválido!")
```
