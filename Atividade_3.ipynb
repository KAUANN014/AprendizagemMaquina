```python
import csv

# 1. Criar o arquivo dados.csv com os novos nomes alterados
dados = [
    ['Nome', 'Idade'],
    ['Neymar', 25],
    ['Vinícius Jr', 30],
    ['Xavi', 22],
    ['Iniesta', 28],
    ['Leonardo', 35]
]

# Criando o arquivo CSV e escrevendo os dados
with open('dados.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(dados)

# 2. Ler o arquivo dados.csv e armazenar os dados em uma lista
pessoas = []
with open('dados.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Ignorar o cabeçalho
    for row in reader:
        nome, idade = row
        pessoas.append([nome, int(idade)])

# Função para verificar se o nome está na lista e mostrar os dados
def verificar_nome(nome_digitado):
    # Verifica se o nome está na lista
    pessoa_encontrada = None
    for pessoa in pessoas:
        if pessoa[0].lower() == nome_digitado.lower():
            pessoa_encontrada = pessoa
            break

    if pessoa_encontrada:
        # Pessoa foi encontrada
        nome, idade = pessoa_encontrada
        # Verificar se é a pessoa mais velha
        mais_velha = max(pessoas, key=lambda p: p[1])
        if pessoa_encontrada == mais_velha:
            print(f"{nome} tem {idade} anos, é a pessoa mais velha da lista.")
        else:
            print(f"{nome} tem {idade} anos, não é a pessoa mais velha da lista.")
    else:
        # Nome não encontrado
        print("Nome não encontrado.")

# Pedir para o usuário digitar um nome
nome_input = input("Digite um nome: ")
verificar_nome(nome_input)
```
