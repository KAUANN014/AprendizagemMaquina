```python
import csv
from datetime import datetime

data = [
    ("Neymar", "05/12/1990", "2025-02-24", "14:30:00"),
    ("Vinicius JR", "15/02/1985", "2025-02-24", "15:00:00"),
    ("João", "20/08/1992", "2025-02-24", "09:45:00"),
    ("Irineu Silva", "11/10/1995", "2025-02-24", "11:15:00"),
    ("Luiz", "30/03/1988", "2025-02-24", "17:00:00"),
]

file_name = "/content/cadastro.csv"

with open(file_name, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Nome", "Data de Nascimento", "Dia de Cadastro", "Hora de Cadastro"])
    writer.writerows(data)

with open("/content/cadastro.csv", mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)
    registros = [row for row in reader]

registro_n = int(input("Digite o número do registro (1 a 5): "))

if 1 <= registro_n <= len(registros):
    nome, data_nascimento, dia_cadastro, hora_cadastro = registros[registro_n - 1]
    dia_formatada = datetime.strptime(dia_cadastro, "%Y-%m-%d").strftime("%d/%m/%Y")
    hora_formatada = datetime.strptime(hora_cadastro, "%H:%M:%S").strftime("%H:%M")
    print(f"Registro {registro_n}: Nome: {nome}; Data de nascimento: {data_nascimento}; Data de cadastro: {dia_formatada} às {hora_formatada} horas")
else:
    print("Número do registro inválido!")
```
