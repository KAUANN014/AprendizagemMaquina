```python
def obter_frase():
    frase = input("Digite uma frase: ")
    if not frase.strip():
        raise ValueError("A frase não pode ser vazia.")
    return frase

def main():
    try:
        frase = obter_frase()
        palavras = frase.split()
        num_caracteres = len(frase)
        num_palavras = len(palavras)
        maior_palavra = max(palavras, key=len)
        frase_invertida_caracteres = frase[::-1]
        frase_invertida_palavras = " ".join(palavras[::-1])
        maiusculas, minusculas = frase.upper(), frase.lower()
        palavras_tupla = tuple(palavras)

        print(f"\nCaracteres: {num_caracteres}")
        print(f"Palavras: {num_palavras}")
        print(f"Maior palavra: {maior_palavra}")
        print(f"Invertida (caracteres): {frase_invertida_caracteres}")
        print(f"Invertida (palavras): {frase_invertida_palavras}")
        print(f"Maiúsculas: {maiusculas}")
        print(f"Minúsculas: {minusculas}")
        print(f"Tupla de palavras: {palavras_tupla}")

    except ValueError as e:
        print(e)

main()
```
