```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

dados = {
    "IMC": [17.2, 18.4, 19.8, 22.0, 24.5, 25.5, 27.8, 30.1, 32.5, 35.0],
    "Obeso": [False, False, False, False, False, False, False, True, True, True]
}

df = pd.DataFrame(dados)
df.to_csv("IMC.csv", index=False)

data = pd.read_csv("IMC.csv")
X = data[["IMC"]]
y = data["Obeso"]

model = DecisionTreeClassifier()
model.fit(X, y)

valor_imc = float(input("Digite um valor de IMC para saber se é obeso (ex: 40.0): "))
novo_imc_df = pd.DataFrame({"IMC": [valor_imc]})
resultado = model.predict(novo_imc_df)

print(f"IMC: {valor_imc} -> Obeso: {resultado[0]}")

```
